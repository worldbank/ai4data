# AI4Data Dataset Mention Extraction Demo

This notebook demonstrates how to extract dataset mentions from text using the `extract_from_text` function in `ai4data`. The function extracts dataset name mentions and automatically classifies their `typology_tag` and `usage_context` using native zero-shot sentence-level classification.

```python
import os
import sys
import json

# Add the src directory to python path if running inside the repository
src_path = os.path.abspath("src")
if src_path not in sys.path:
    sys.path.insert(0, src_path)

from ai4data import extract_from_text
```

## 1. Define Example Text

```python
text = (
    "To analyze the impact of infrastructure spillovers, we combine data from the "
    "2010 Ghana Living Standards Survey (GLSS) with production records for 17 "
    "large-scale gold mines."
)
print("Input Text:\n", text)
```

## 2. Run Dataset Mention Extraction

We run the extraction with `include_confidence=True` and pass the fine-tuned adapter ID `ai4data/datause-extraction`.

```python
result = extract_from_text(
    text,
    include_confidence=True,
    adapter_id="ai4data/datause-extraction"
)
```

## 3. Print Extracted Mentions

```python
for ds in result.get("datasets", []):
    print(f"Dataset Mention: {ds['mention_name']['text']}")
    print(f"  Confidence:    {ds['mention_name']['confidence']:.3f}")
    print(f"  Typology Tag:  {ds['typology_tag']['text']} (confidence: {ds['typology_tag']['confidence']:.3f})")
    print(f"  Usage Context: {ds['usage_context']['text']} (confidence: {ds['usage_context']['confidence']:.3f})")
    print(f"  Is Used:       {ds['is_used']['text']} (confidence: {ds['is_used']['confidence']:.3f})")
    print("-" * 60)
```

## 4. Full JSON Output

```python
print(json.dumps(result, indent=2))
```
