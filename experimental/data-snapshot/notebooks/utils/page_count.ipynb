```python
from pathlib import Path
import pymupdf

def count_total_page_count(doc_paths):
    page_count = 0
    for f in doc_paths:
        with pymupdf.open(f) as doc:
            page_count += doc.page_count
    return page_count
```

```python
unhcr = list(Path("../pdf_input/").rglob("unhcr_batch*/*.pdf"))
page_count = count_total_page_count(unhcr)

print(len(unhcr))
print(page_count)
```
