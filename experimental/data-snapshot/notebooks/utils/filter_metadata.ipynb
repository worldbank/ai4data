```python
import glob
from pathlib import Path
import shutil
from tqdm.auto import tqdm
```

```python
pdfs = [Path(x) for x in glob.glob("../../pdf_input/prwp_batch1/*.pdf")] + [Path(x) for x in glob.glob("../../pdf_input/prwp_batch2/*.pdf")]
len(pdfs)
```

```python
metadata = [Path(x) for x in glob.glob("/home/ajd/corpus/prwp/metadata/*.json")]
len(metadata)
```

```python
for p in tqdm(pdfs):
    to_copy = [m for m in metadata if m.stem.replace("_metadata", "") == p.stem]

    if to_copy:
        for c in to_copy:
            shutil.copy(c, Path("../../data/hf_metadata/prwp_original/") / c.name)
    else:
        print(f"No metadata found for {p.name}")
```
