```python
import glob
import math
from pathlib import Path
import pymupdf
import random
import shutil
from tqdm.auto import tqdm
```

```python
# Inputs

## Source dir
# SOURCE = "prwp"
SOURCE = "refugee"
SOURCE_DIR = f"../../corpus/{SOURCE}/pdfs/"

## Save dir
SAVE_DIR = f"annotation_corpus/"

## List files to exclude for sampling
exc_lst = []
exc_glob = [
    # "annotation_corpus/unhcr_batch*/*.pdf",
]

## Sampling
BATCH_START = 1
BATCH_END = 1
SAMPLE_SIZE = 50
# MAX_PAGE_COUNT = 40
MAX_PAGE_COUNT = 60
```

# Preliminary steps

```python
SAVE_DIR = Path(SAVE_DIR)
SOURCE_DIR = Path(SOURCE_DIR)
```

```python
# Exclude files
for g in exc_glob:
    exc_lst += [Path(x).stem for x in glob.glob(g)]

len(exc_lst)
```

```python
files = list(SOURCE_DIR.rglob("*.pdf"))
len(files)
```

```python
files = [x for x in files if x.stem not in exc_lst]
len(files)
```

```python
# Filter long documents
short_files = []
for f in tqdm(files):
    try:
        with pymupdf.open(f) as doc:
            if doc.page_count <= MAX_PAGE_COUNT:
                short_files.append(f)
    except Exception as e:
        print(e)

len(short_files)
```

# Sampling

```python
shortlist = [x for x in short_files]
for b in range(BATCH_START, BATCH_END + 1, 1):
    if len(shortlist) > SAMPLE_SIZE:
        sample = random.sample(shortlist, SAMPLE_SIZE)
    else:
        sample = [x for x in shortlist]

    # Copy files
    for s in sample:
        save_dir_ = SAVE_DIR / f"{SOURCE}_batch{b}"
        save_dir_.mkdir(parents=True, exist_ok=True)
        shutil.copy2(s, save_dir_ / s.name)
    
    # Remove already sampled files from shortlist
    shortlist = [x for x in shortlist if x not in sample]
    print(len(set(shortlist)))
```
