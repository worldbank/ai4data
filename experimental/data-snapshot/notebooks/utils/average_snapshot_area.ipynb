```python
from dsa.constants import ROOT
from dsa.evaluate_model import bbox_area
from dsa.utils import load_json
```

```python
j = load_json(ROOT / "data/paper/evaluation_input/ground_truth_corpus_combined.json")["predictions"]
```

```python
obj = {}  # page_id: list of objects

for p in j:
    obj[p["page_id"]] = p["objects"]
```

```python
areas = []

for _, bboxes in obj.items():
    area = 0
    for b in bboxes:
        area += bbox_area(b["bbox"])

    areas.append(area)
```

```python
import numpy as np
```

```python
np.mean(areas)
```

```python
np.median(areas)
```
