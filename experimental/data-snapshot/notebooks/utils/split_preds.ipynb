```python
import json
from pathlib import Path
from tqdm.auto import tqdm

from dsa.utils import load_json
```

```python
gt = load_json("../data/evaluation_input/ground_truth_unhcr_combined.json")
```

```python
gt.keys()
```

```python
label_map = gt["label_map"]
info = gt["info"]
documents = gt["documents"]
predictions = gt["predictions"]
```

```python
for doc in tqdm(documents):
    # Create new dict
    new_json = {}
    new_json["label_map"] = label_map
    new_json["info"] = info
    new_json["documents"] = doc
    new_json["predictions"] = []

    # Keep only predictions for the document
    doc_id = doc["doc_id"]
    for preds in predictions:
        if preds["doc_id"] == doc_id:
            new_json["predictions"].append(preds)

    if not new_json["predictions"]:
        print(f"{doc_id} has no annotations.")

    # Save 1 annotation file per document
    fname = str(Path(doc_id).stem)
    with open(f"data/split_preds/{fname}_annotations.json", "w") as f:
        json.dump(new_json, f, indent=2)
```
