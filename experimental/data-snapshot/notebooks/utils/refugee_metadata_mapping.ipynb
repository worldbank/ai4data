```python
from dsa.utils import load_json
import glob
import json
from openpyxl import load_workbook
import os
```

```python
def find_guid(j):

    for x in j["document_description"]["identifiers"]:
        if x["type"] == "World Bank GUID":
            return x["identifier"]
    print(f"No GUID found for {j['metadata_information']['idno']}")
    return ""
```

# Create mapping: idno, fname, guid

```python
files = glob.glob("../../corpus/refugee/metadata/*.json")
len(files)
```

```python
map1 = {}  # {document_{idno}.json: [fname.pdf, guid]}
for f in files:
    key = os.path.basename(f)

    j = load_json(f)
    fname = os.path.basename(j["document_description"]["url"])
    guid = find_guid(j)

    map1[key] = [fname, guid]

map1
```

# Create mapping: idx, guid

```python
wb = load_workbook(
    "../../corpus/refugee/Results Data_project_docs_refugee.xlsx", data_only=True
)
ws = wb["data"]
url_list = [
    cell.hyperlink.target
    for row in ws.iter_rows(min_row=2)
    if (cell := row[0]).hyperlink
]
guid_list = [x.split("/")[-1] for x in url_list]

map2 = {f'{i:03d}': x for i, x in enumerate(guid_list, start=1)}
map2
```

# Create mapping: idx, fname

```python
indexed_fnames = [os.path.basename(x) for x in glob.glob("../../corpus/refugee/pdfs/*.pdf")]
map3 = {x[:3]:x for x in indexed_fnames}

map3
```

# Create master table

```python
import pandas as pd
```

```python
map1_df = (
    pd.DataFrame.from_dict(map1, orient="index", columns=["fname", "guid"])
    .reset_index()
    .rename(columns={"index": "metadata"})
)
map2_df = pd.DataFrame(list(map2.items()), columns=["idx", "guid"])
map3_df = pd.DataFrame(list(map3.items()), columns=["idx", "indexed_fname"])

df = map1_df.merge(map2_df, how="left", on="guid").merge(map3_df, how="left", on="idx") 
df
```

# Copy and rename metadata

```python
import shutil
from pathlib import Path
```

```python
metadata_map = df.set_index("indexed_fname")["metadata"].to_dict()
metadata_map
```

```python
annot_corpus = glob.glob("annotation_corpus/refugee_batch1/*.pdf")

metadata_dir = "../../corpus/refugee/metadata/"
metadata_renamed_dir = "../../corpus/refugee/metadata_indexed_fname/"

for f in annot_corpus:
    ifn = os.path.basename(f)
    meta = metadata_map[ifn]
    
    to_copy = metadata_dir + meta
    dest = metadata_renamed_dir + (Path(ifn).stem + ".json")
    shutil.copy(to_copy, dest)
```
