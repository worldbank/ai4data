```python
sample_size = 5
for d in dirs:
    files = list(d.glob("*.png"))
    samples = random.sample(files, k=sample_size)

    # Copy samples
    out_path = Path(f"../data/visualize_pages/{d.stem}/")
    out_path.mkdir(parents=True, exist_ok=True)
    for s in samples:
        copy(s, out_path / s.name)
    
```
