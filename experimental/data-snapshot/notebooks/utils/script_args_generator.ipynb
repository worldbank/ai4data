```python
from pathlib import Path
from dsa.constants import ROOT
```

```python
# text = "python src/scripts/visualize_pages.py \
# --gt_json_path=data/paper/evaluation_input/gt/ground_truth_refugee_combined.json \
# --pred_json_path=data/paper/evaluation_input/doclayoutyolo/doclayoutyolo_refugee_batch1_wo_small.json \
# --input_pdf_dir=../hf_datasets/data-snapshot/documents/refugee/ \
# --output_dir=data/visualize_pages/doclayoutyolo_refugee"

# text = f"python src/scripts/visualize_pages.py \
# --gt_json_path=data/paper/evaluation_input/{gt} \
# --pred_json_path=data/paper/evaluation_input/{model}/{pred} \
# --input_pdf_dir=../hf_datasets/data-snapshot/documents/{source}/ \
# --output_dir=data/visualize_pages/{model}_{source}/"

models = [
    "tfid-large",
    "doclayoutyolo",
    "yolo11",
    "yolo26",
]
sources = [
    "unhcr",
    "prwp",
    "refugee",
]
gt_lookup = {x: f"gt/ground_truth_{x}_combined.json" for x in sources}
params = []  # list of (gt_path, pred_path, input_pdf_dir, output_dir)

for m in models:
    for s in sources:
        gt = f"data/paper/evaluation_input/{gt_lookup[s]}"
        pred = f"data/paper/evaluation_input/{m}/{m}_{s}_combined_wo_small.json"
        pdf_dir = f"../hf_datasets/data-snapshot/documents/{s}"
        output_dir = f"data/visualize_pages/refugee_pred/{m}_{s}/"
        params.append((gt, pred, pdf_dir, output_dir))
```

```python
# Check if files exist
for p in params:
    for x in p:
        if Path(ROOT / x).exists() is False:
            print(f"Does not exist: {x}")
```

```python
text = "python src/scripts/visualize_pages.py \
--gt_json_path={} \
--pred_json_path={} \
--input_pdf_dir={} \
--output_dir={} \
--mode='pred'"

for p in params:
    text2 = text.format(*p)
    print(text2)
    print("&&")
```
