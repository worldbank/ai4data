```python
from dsa.constants import ROOT
from dsa.utils import load_json
import pandas as pd
```

```python
res_dir = ROOT / "data/paper/evaluation_output/"
res_files = sorted(list(res_dir.glob("report_*.json")))

res_files
```

```python
# for f in res_files:
#     f.resolve(strict=True)
```

```python
thresholds = [
    "0.5",
    # "0.75",
]
classes = [
    "Figure",
    "Table",
]
metrics = {
    "precision": "detection",
    "recall": "detection",
    "mean_iou": "spatial accuracy",
    "mean_area_precision": "spatial accuracy",
    "mean_area_recall": "spatial accuracy",
}
```

```python
cols = []
for r in res_files:
    row = {}
    row["filename"] = r.name

    res = load_json(r)
    for t in thresholds:
        for c in classes:
            for m in metrics:
                row[f"[{c}] {m} @ {t}"] = res["metrics"][t]["per_class"][c][m]

    cols.append(row)
```

```python
df = pd.DataFrame(cols).set_index("filename").transpose()
df
```
