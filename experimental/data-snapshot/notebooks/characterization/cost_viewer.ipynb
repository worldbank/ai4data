```python
import glob
import pandas as pd
```

```python
files = glob.glob("snapshot_to_shortdesc//data/results_*.jsonl")
files
```

```python
df = pd.concat([pd.read_json(x, lines=True) for x in files])
df
```

```python
df = df.join(pd.json_normalize(df["cost"]))
```

```python
df["total_cost_usd"].sum()
```

```python
df["total_cost_usd"].mean()
```

```python
df["total_cost_usd"].std()
```

```python
df.groupby("source_label").agg(
    count=("source_label", "count"),
    total_cost_usd=("total_cost_usd", "sum"),
)
```
