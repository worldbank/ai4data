# Create project and tasks

```python
import os
from dotenv import load_dotenv
from label_studio_sdk import LabelStudio
from dsa.constants import ROOT
import json

load_dotenv()
API_KEY = os.getenv("LABELSTUDIO_API_KEY")

# Load labeling template
with open(ROOT / "labeling_interface_template.xml", "r", encoding="utf-8") as f:
    label_config = f.read()

# Initialize the client
client = LabelStudio(base_url="http://localhost:8080", api_key=API_KEY)

# Create the project
## Input parameter: project_name
project_name = "programmatic7"
project = client.projects.create(title=project_name, label_config=label_config)
print(f"Project created with ID: {project.id}")

# Create the json task file with prelabeling
## Input parameter: dataset_folder
## Input parameter: pred_json_file
dataset_folder = ROOT / "labelstudio_data/unhcr_batch2"
with open(dataset_folder / "tasks_prelabeled.json", "r") as f:
    payload = json.load(f)

# Import task to the project
client.projects.import_tasks(id=project.id, request=payload)
```

# Export project

```python
import os
from dotenv import load_dotenv
from label_studio_sdk import LabelStudio

load_dotenv()
API_KEY = os.getenv("LABELSTUDIO_API_KEY")

# Initialize the client
client = LabelStudio(base_url="http://localhost:8080", api_key=API_KEY)

response = client.projects.exports.download_sync(
    id=22,
    download_all_tasks=True,
    export_type="JSON",
)

decoded = json.loads(b"".join(response).decode())

# Export
with open('test.json', 'w') as f:
    json.dump(decoded, f, indent=2)
```
